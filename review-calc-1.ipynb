| Indeterminate Form | How to Rewrite | Example Setup | Result after Rewriting |
|--------------------|----------------|---------------|------------------------|
| $\tfrac{0}{0} $ | Already in quotient form | $lim_{x \to 0} \frac{\sin x}{x}$ | Apply L'Hôpital directly |
| $\tfrac{\infty}{\infty} $ | Already in quotient form | $\lim_{x \to \infty} \frac{x}{e^x}$ | Apply L'Hôpital directly |
| $0 \cdot \infty$ | Rewrite as $\frac{\ln x}{1/x}$ or similar | $\lim_{x \to 0^+} x \ln x$ | Becomes $\tfrac{\infty}{\infty}$ |
| $\infty - \infty$ | Combine into single fraction (e.g. conjugate) | $\lim_{x \to \infty} (\sqrt{x^2+1} - x)$ | Becomes $\tfrac{0}{0}$ |
| $0^0$ | Take logs: $g(x)\ln f(x)$ | $\lim_{x \to 0^+} x^x$ | Reduces to $0 \cdot \infty$ |
| $\infty^0$ | Take logs: $g(x)\ln f(x)$ | $\lim_{x \to \infty} (1+1/x)^x$ | Reduces to $\tfrac{0}{0}$ |
| $1^\infty$ | Take logs: $g(x)\ln f(x)$ | $\lim_{x \to \infty} (1+1/x)^x$ | Reduces to $\tfrac{0}{0}$ |
